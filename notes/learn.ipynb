## 実際に表情を学習

```python
import numpy as np
import pandas as pd
import json
from tqdm import tqdm
```

```python
with open('../data/lean_data.csv', 'r') as f:
    data = pd.read_csv(f)
data.dropna()
```

```python
def get_feature_flatten(features_str):
    features = json.loads(features_str)
    features = [np.array(feature).flatten() for feature in features]
    features = [value for feature in features for value in feature]
    return features
```

```python
x = data['features_norm']
x = list(map(get_feature_flatten, tqdm(x.values)))
x = np.array(x)
x[np.isnan(x)] = 0
x.shape
```

```python
y = data['label'].values
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
forest = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=1)
```

```python
forest.fit(x_train, y_train)
```

```python
y_pred = forest.predict(x_test)
```

```python
accuracy_score(y_true=y_test, y_pred=y_pred)
```
