## FERデータセットの中身を確認

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import face_recognition
```

```python
with open('../data/fer2013.csv', 'r') as f:
    data = pd.read_csv(f)
data
```

```python
pixel_test = np.array([pixel for pixel in data['pixels'][7].split(' ')])
pixel_test = pixel_test.astype('uint8')
pixel_test = pixel_test.reshape(48, 48)
pixel_test_rgb = cv2.cvtColor(pixel_test, cv2.COLOR_GRAY2RGB)
```

```python
plt.imshow(pixel_test)
plt.show()
```

```python
face_landmarks_list = face_recognition.face_landmarks(pixel_test_rgb)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.imshow(pixel_test_rgb)
for parts_name, parts_pos in face_landmarks_list[0].items():
    parts = np.array([[pos[0], pos[1]] for pos in parts_pos])
    ax.scatter(parts[:, 0], parts[:, 1], s=20)
# plt.savefig('../data/output_obama2_parts.jpg')
plt.show()
```
